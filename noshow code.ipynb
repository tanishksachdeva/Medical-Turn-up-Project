```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv('KaggleV2-May-2016.csv')
```

```python
df.head()
```

```python
df.sort_values(['AppointmentDay', 'ScheduledDay']  , ascending=True , inplace= True)
df.head(10)
```

```python
df.rename(columns={'PatientId':'PatientID', 'Neighbourhood':'Neighborhood',
                   'Hipertension':'Hypertension', 'Handcap':'Handicap',
                   'No-show':'NoShow'}, inplace=True)
```

```python
df.head()
```

```python
print('nos of pateints ' , df.shape[0])
print('nos of unique pateints ' , len(df.PatientID.unique()))
print('nos of app id unique ' , len(df.AppointmentID.unique()))
```

```python
print('Scholarship:\t%s' %df.Scholarship.unique())
print('Hypertension:\t%s' %df.Hypertension.unique())
print('Diabetes:\t%s' %df.Diabetes.unique())
print('Alchoholism:\t%s' %df.Alcoholism.unique())
print('Handicap:\t%s' %df.Handicap.unique())
print('SMS_received:\t%s' %df.SMS_received.unique())
print('NoShow:\t%s' %df.NoShow.unique())
```

```python
df.drop(df[(df.Age < 0) | (df.Age > 100)].index, inplace = True)
```

```python
len(df.Age)
```

```python
df.ScheduledDay=df.ScheduledDay.apply(np.datetime64)
df.AppointmentDay=df.AppointmentDay.apply(np.datetime64)
```

```python
df.head()
```

```python
df['Time'] = df['ScheduledDay'].sub(df['AppointmentDay'] , axis = 0)
```

```python
df.head()

```

```python
df['Time'] = df['Time'] / np.timedelta64(1, 'D')
df['Time']=df['Time'].abs().apply(np.floor)
```

```python
df['NoShow'] = pd.get_dummies(df['NoShow'])
```

```python
df.head()
```

```python
df['Handicap'] = pd.Categorical(df['Handicap'])
df_Handicap = pd.get_dummies(df['Handicap'], prefix = 'Handicap')
df = pd.concat([df, df_Handicap], axis=1)
```

```python
df['Neighborhood'] = pd.Categorical(df['Neighborhood'])
df['Neighborhood'] = df['Neighborhood'].cat.codes
df_Neighborhood = pd.get_dummies(df['Neighborhood'], prefix = 'Neighborhood')
df = pd.concat([df, df_Neighborhood], axis=1)
```

```python
df['Gender'] = pd.get_dummies(df['Gender'])
```

```python
df['prev_appMissed'] = df.groupby('PatientID')['NoShow'].apply(lambda x: x.shift(-1) == 1)
df.prev_appMissed = df.prev_appMissed.astype(int)

# New feature = number of appointments missed by patient
df['n_appMissed'] = df.groupby('PatientID')['NoShow'].apply(lambda x: x.cumsum())
```

```python
df.describe()
```

```python
df.drop(['PatientID','AppointmentID','ScheduledDay',
         'AppointmentDay','Neighborhood', 'Handicap'], axis=1, inplace=True)
```

```python
cols = df.columns.tolist()
cols[0:10] = ['NoShow','Age','Time','Gender','Scholarship','Hypertension','Diabetes',
              'Alcoholism','SMS_received']
df = df[cols].astype(float)
df = pd.DataFrame.sort_index(df)
df.head()
```

```python
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.externals import joblib
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['NoShow' ,'Neighborhood_80' , 'Handicap_4'],axis=1)
y = df['NoShow']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)

```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
 rfc= RandomForestClassifier(n_estimators=1000)
```

```python
rfc.fit(X_train,y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```
